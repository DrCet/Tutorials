````python
import pandas as pd
import numpy as np
````

###**1. Create pd.Series**


```
series = pd.Series(data = [list of value], index = [list of keys], name = 'String name')
```

See more [here](https://pandas.pydata.org/docs/reference/api/pandas.Series.html)



````python
data = [12, 24, 134]
index = ['a', 'b','c']
name = 'My First Series'
my_series = pd.Series(data = data, index = index, name = name, dtype = 'int64')
my_series
````

###**2. Basic selection**
There are many ways we can try:

```
my_series['key' or list of keys]
```
```
my_series.loc['key' or list of keys]
```
```
my_seroes.iloc[index or list of indexes]

___________________________________________________________

````python
second_to_last = my_series.iloc[-2]
print('Second to last value:',second_to_last)
first_value = my_series.loc['a']
print('Value of a:', first_value)
more_values = my_series.loc[['a', 'b', 'c']]
print('Select multiple values:', more_values)
````

###**5. Sorting**
Sorting will set `inplace = False` by default so that it will return a series

To modify the series directly set `inplace = True` *(not recommended)*

````python
my_series.sort_values(ascending = False)
````

````python
my_series.sort_index(ascending = False)
````

````python
my_series.sort_values(ascending = False,
                      inplace = True)
````
